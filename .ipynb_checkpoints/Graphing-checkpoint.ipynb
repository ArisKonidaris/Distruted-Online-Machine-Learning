```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the measurments from the csv file

```python
file="Graphs/Distr_Learn_CommTest.csv"

info=['NetName', 'DistributedAlgo', 'Learner']
measurements=['Sites', 'Rounds', 'Subrounds', 'Safezones', \
              'TotalMsgs', 'TotalInfoMsgs', 'TotalBytes', \
              'Accuracy', 'Threshold', 'BatchSize']

# Loading the experiments.
data = pd.DataFrame(pd.read_csv(file))
data
```

# Accuracy vs Number of Sites for various Thresholds

```python
## Accuracy vs Number of Sites for Batch Size = 192 ##
acc_sit_thr_192=data.query('BatchSize==192')[['Accuracy','Sites','Threshold']]
plt1=acc_sit_thr_192.query('Threshold==0.0008')[['Accuracy','Sites']].values
plt2=acc_sit_thr_192.query('Threshold==0.008')[['Accuracy','Sites']].values
plt3=acc_sit_thr_192.query('Threshold==0.08')[['Accuracy','Sites']].values

plt.plot(plt1[:,1],plt1[:,0],'o-',linewidth=2)
plt.plot(plt2[:,1],plt2[:,0],'o-',linewidth=2)
plt.plot(plt3[:,1],plt3[:,0],'o-',linewidth=2)
plt.title('Accuracy vs Sites for BatchSize = 192')
plt.xlabel('Number of Sites')
plt.ylabel('Accuracy')
plt.legend(['Threshold : 0.0008','Threshold : 0.008','Threshold : 0.08'])
plt.grid()
plt.show()


## Accuracy vs Number of Sites for Batch Size = 128 ##
acc_sit_thr_128=data.query('BatchSize==128')[['Accuracy','Sites','Threshold']]
plt1=acc_sit_thr_128.query('Threshold==0.0008')[['Accuracy','Sites']].values
plt2=acc_sit_thr_128.query('Threshold==0.008')[['Accuracy','Sites']].values
plt3=acc_sit_thr_128.query('Threshold==0.08')[['Accuracy','Sites']].values

plt.plot(plt1[:,1],plt1[:,0],'o-',linewidth=2)
plt.plot(plt2[:,1],plt2[:,0],'o-',linewidth=2)
plt.plot(plt3[:,1],plt3[:,0],'o-',linewidth=2)
plt.title('Accuracy vs Sites for BatchSize = 128')
plt.xlabel('Number of Sites')
plt.ylabel('Accuracy')
plt.legend(['Threshold : 0.0008','Threshold : 0.008','Threshold : 0.08'])
plt.grid()
plt.show()


## Accuracy vs Number of Sites for Batch Size = 64 ##
acc_sit_thr_64=data.query('BatchSize==64')[['Accuracy','Sites','Threshold']]
plt1=acc_sit_thr_64.query('Threshold==0.0008')[['Accuracy','Sites']].values
plt2=acc_sit_thr_64.query('Threshold==0.008')[['Accuracy','Sites']].values
plt3=acc_sit_thr_64.query('Threshold==0.08')[['Accuracy','Sites']].values

plt.plot(plt1[:,1],plt1[:,0],'o-',linewidth=2)
plt.plot(plt2[:,1],plt2[:,0],'o-',linewidth=2)
plt.plot(plt3[:,1],plt3[:,0],'o-',linewidth=2)
plt.title('Accuracy vs Sites for BatchSize = 64')
plt.xlabel('Number of Sites')
plt.ylabel('Accuracy')
plt.legend(['Threshold : 0.0008','Threshold : 0.008','Threshold : 0.08'])
plt.grid()
plt.show()
```
